```python
import numpy as np
import matplotlib.pyplot as plt
from surfaces.procedural_surface import procedural_surface
```

```python
K = np.linspace(0.7, 1.3, 120)
T = np.linspace(0.05, 2.0, 120)

KK, TT = np.meshgrid(K, T)
IV = procedural_surface(KK, TT, seed=7)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(KK, TT, IV)
ax.set_xlabel("Moneyness K/S")
ax.set_ylabel("Maturity T")
ax.set_zlabel("Implied Volatility")
ax.set_title("Procedural Volatility Terrain with Crash Caves")

plt.show()
```
